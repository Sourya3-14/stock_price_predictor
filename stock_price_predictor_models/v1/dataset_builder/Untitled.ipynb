```python
import pandas as pd
import numpy as np
import ast
```

```python
# Load datasets
df_0816 = pd.read_csv("dataset/global_news_08-16.csv")
```

```python
# convert date column
df_0816["Date"] = pd.to_datetime(df_0816["Date"])

# select news columns
news_cols = [c for c in df_0816.columns if c.startswith("Top")]

# melt columns into rows
df_long = df_0816.melt(
    id_vars="Date",
    value_vars=news_cols,
    value_name="news"
)

# keep only required columns
df_long = df_long[["Date", "news"]]

# rename
df_long = df_long.rename(columns={"Date": "date"})
```

```python
# convert byte strings -> normal strings
def decode_news(x):
    if isinstance(x, str) and x.startswith("b'") or isinstance(x, str) and x.startswith('b"'):
        return ast.literal_eval(x).decode("utf-8", errors="ignore")
    return x
```

```python
df_long["news"] = df_long["news"].apply(decode_news)

# remove empty rows
df_long = df_long.dropna(subset=["news"])

# sort by date
df_long = df_long.sort_values("date")

# reset index
df_long = df_long.reset_index(drop=True)
df_0816 = df_long.dropna()
```

```python
df_0920 = pd.read_csv("dataset/news_09-20.csv")
```

```python
# keep only required columns
df_0920 = df_0920[["headline", "date"]]

# convert to datetime
df_0920["date"] = pd.to_datetime(df_0920["date"], errors="coerce")

# remove timezone if present
df_0920["date"] = df_0920["date"].dt.tz_localize(None)

# keep only date (remove time)
df_0920["date"] = df_0920["date"].dt.date

# rename column
df_0920 = df_0920.rename(columns={"headline": "news"})

# sort by date
df_0920 = df_0920.sort_values("date").reset_index(drop=True)

df_0920 = df_0920.dropna()
```

```python
df_1624 = pd.read_csv("dataset/apple_news_16-24.csv")
```

```python
# create news column (prefer content, fallback to title)
df_1624["news"] = df_1624["content"].fillna(df_1624["title"])

# keep only required columns
df_1624 = df_1624[["date", "news"]]

# convert to datetime
df_1624["date"] = pd.to_datetime(df_1624["date"], errors="coerce")

# remove timezone
df_1624["date"] = df_1624["date"].dt.tz_localize(None)

# keep only date
df_1624["date"] = df_1624["date"].dt.date

# drop missing
df_1624 = df_1624.dropna(subset=["news"])

# sort
df_1624 = df_1624.sort_values("date").reset_index(drop=True)
```

```python
df_25 = pd.read_csv("dataset/fin_news_25.csv")
```

```python
# keep relevant columns
df_25 = df_25[["Date", "Headline"]]

# rename columns
df_25 = df_25.rename(columns={
    "Date": "date",
    "Headline": "news"
})

# convert to datetime
df_25["date"] = pd.to_datetime(df_25["date"], errors="coerce")

# keep only date (remove time if present)
df_25["date"] = df_25["date"].dt.date

# drop missing rows
df_25 = df_25.dropna(subset=["news"])

# sort by date
df_25 = df_25.sort_values("date").reset_index(drop=True)
```

```python
df_news = pd.read_csv("dataset/news_dataset.csv")
```

```python
df_news["date"] = pd.to_datetime(df_news["date"], errors="coerce")
df_news = df_news.rename(columns={"headline":"news"})
```

```python
df_news
```

```python
import requests
from dotenv import load_dotenv
import os

load_dotenv()

def fetch_news_for_day(date):
    url = "https://newsapi.org/v2/everything"

    params = {
        "q": "Microsoft OR MSFT",
        "from": date,
        "to": date,
        "language": "en",
        "sortBy": "relevancy",
        "apiKey": os.getenv("NEWS_API")
    }

    r = requests.get(url, params=params)
    data = r.json()

    if data.get("status") != "ok":
        return []

    articles = data.get("articles", [])

    rows = []
    for article in articles:
        rows.append({
            "date": article["publishedAt"].split("T")[0],
            "news": article["title"],
        })

    return rows
```

```python
from datetime import datetime, timedelta

start_date = datetime(2026, 4, 19)
end_date = datetime(2026, 4, 30)

all_data = []

current = start_date

while current <= end_date:
    date_str = current.strftime("%Y-%m-%d")
    print("Fetching:", date_str)

    try:
        daily_news = fetch_news_for_day(date_str)
        all_data.extend(daily_news)

    except Exception as e:
        print("Error on", date_str, e)

    current += timedelta(days=1)
```

```python
df = pd.DataFrame(all_data)
df = df.dropna()
df = df.drop_duplicates()
df["date"] = pd.to_datetime(df["date"], errors="coerce")
```

```python
df_final = pd.concat([df_0816,df_0920,df_1624,df_25,df_news,df], ignore_index=True)
```

```python
df_final["date"] = pd.to_datetime(df_final["date"]).dt.date
```

```python
df_final = df_final.dropna()
df_final = df_final.drop_duplicates()
```
